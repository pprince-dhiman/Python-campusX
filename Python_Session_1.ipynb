## Type conversion
- type conversion (casting) means **creating a new object** of a different type from an existing value.
- It does **NOT change the original** variable automatically.


```python
a = input('Enter val1: ')
b = input('Enter val2: ')
result = int(a)+int(b)
print("Sum: ", result)

print(type(a), type(b), type(result))
```

```python
str(3) # becomes string
val4 = float('4') # becomes 4.0

print('type of 3: ',type(str(3)))
print('type of 4: ',type(float('4')))
print('val4: ', val4)
```

```python
# complex to int
# int(3+4j) -> error (bcz info loss for imaginary part)
realPart = (3+4j).real
imgPart = (3+4j).imag
print('realPart: ', realPart)
print('imgPart: ', imgPart)
```
